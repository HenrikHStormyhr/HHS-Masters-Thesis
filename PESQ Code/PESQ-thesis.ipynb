```python
# Testing PESQ with Steghide Sample 1

# This code is taken from the PESQ pip page (only the paths have been changed):
# https://pypi.org/project/pesq/

from scipy.io import wavfile
from pesq import pesq

rate, ref = wavfile.read("./audio/stegref_Female1_Sample1.wav")
rate, deg = wavfile.read("./audio/steghide_Female1_Sample1.wav")

# We will use wideband since our files are 16000hz sample rate.
print(pesq(rate, ref, deg, 'wb'))

# Narrowband is for 8000hz sample rate and will not be used.
# print(pesq(rate, ref, deg, 'nb'))
```

```python
import os
```

```python
from csv import writer
```

```python
# Steghide

# Steghide PESQ calculation added to CSV file.
# Microsoft CoPilot was used to help modify the code.

# Creating CSV File
# This tutorial was followed: https://www.geeksforgeeks.org/how-to-append-a-new-row-to-an-existing-csv-file/

field_names = ['File Name', 'PESQ Score']

with open('./results/steghide_pesq_results.csv', 'a', newline='') as f_object:
 
    # Pass this file object to csv.writer()
    # and get a writer object
    writer_object = writer(f_object)
 
    # Pass the list as an argument into
    # the writerow()
    writer_object.writerow(field_names)
 
    # Close the file object
    #f_object.close()

for x in os.scandir("./audio/steg4pgpref16hz"):
    #print(str(x.path))
    #print(str(x.path).replace("./audio/stegref", ".audio/steghide"))

    rate, ref = wavfile.read(str(x.path))
    rate, deg = wavfile.read(str(x.path).replace("./audio/steg4pgpref16hz", "./audio/steghide16hz"))

    # print(pesq(rate, ref, deg, 'wb'))
    csv_row_list = [str(x.path).replace("./audio/steg4pgpref16hz", ""), pesq(rate, ref, deg, 'wb')]

    with open('./results/steghide_pesq_results.csv', 'a', newline='') as f_object:
 
        # Pass this file object to csv.writer()
        # and get a writer object
        writer_object = writer(f_object)
 
        # Pass the list as an argument into
        # the writerow()
        writer_object.writerow(csv_row_list)
 
# Close the file object
f_object.close()
```

```python
# Hide4PGP

# Hide4PGP PESQ calculation added to CSV file.
# Microsoft CoPilot was used to help modify the code.

# Creating CSV File
# This tutorial was followed: https://www.geeksforgeeks.org/how-to-append-a-new-row-to-an-existing-csv-file/

field_names = ['File Name', 'PESQ Score']

# This CSV path needs to be changed for each method.
with open('./results/hide4pgp_pesq_results.csv', 'a', newline='') as f_object:
 
    # Pass this file object to csv.writer()
    # and get a writer object
    writer_object = writer(f_object)
 
    # Pass the list as an argument into
    # the writerow()
    writer_object.writerow(field_names)
 
    # Close the file object
    #f_object.close()

# The path here needa to be changed for each stego method
for x in os.scandir("./audio/steg4pgpref16hz"):
    #print(str(x.path))
    #print(str(x.path).replace("./audio/stegref", ".audio/steghide"))

    rate, ref = wavfile.read(str(x.path))
    # The paths here needs to be changed for each stego method
    rate, deg = wavfile.read(str(x.path).replace("./audio/steg4pgpref16hz", "./audio/hide4pgp16hz"))

    # print(pesq(rate, ref, deg, 'wb'))
    # The path here needs to be changed for each stego method
    csv_row_list = [str(x.path).replace("./audio/steg4pgpref16hz", ""), pesq(rate, ref, deg, 'wb')]

    # This CSV path needs to be changed for each method.
    with open('./results/hide4pgp_pesq_results.csv', 'a', newline='') as f_object:
 
        # Pass this file object to csv.writer()
        # and get a writer object
        writer_object = writer(f_object)
 
        # Pass the list as an argument into
        # the writerow()
        writer_object.writerow(csv_row_list)
 
# Close the file object
f_object.close()
```
